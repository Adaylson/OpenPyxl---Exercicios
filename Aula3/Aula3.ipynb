# Acessando dados de uma planilha 

```python
from openpyxl import load_workbook

wb = load_workbook('SÃ³ alegria temp.xlsx')

ws = wb['Segunda planilha']
```

```python
for linha in ws.values:
    print(linha)
    for valor in linha:
        print(valor)
```

```python
for row in ws.iter_rows(min_row=1,max_col=9,max_row=10, values_only=True):
    print(row)
```
